```python
from surprise import Dataset
from surprise import SVD
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
data = Dataset.load_builtin('ml-100k')
```

```python
trainset, testset = train_test_split(data, test_size=0.25)
```

```python
model = SVD()
```

```python
model.fit(trainset)
```

```python
predictions = model.test(testset)
```

```python
rmse_score = accuracy.rmse(predictions)
```

```python
rmse_score
```

```python
user_id = str(196)
item_id = str(302)
actual_rating = 4
```

```python
single_prediction = model.predict(user_id, item_id, r_ui=actual_rating)
```

```python
user_id
print(f"Əslində verdiyi reytinq: {single_prediction.r_ui}")
print(f"Modelin təxmin etdiyi reytinq: {round(single_prediction.est, 2)}")
```
